```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

print('Done')
```

```python
np.random.seed(205)

N = 10**5
x = 0
x_values = []
for i in range(N):
  x_rand = np.random.normal(x, scale = 1.0)
  ratio = stats.norm.pdf(x_rand, loc = 2, scale = np.sqrt(2)) / stats.norm.pdf(x, loc = 2, scale = np.sqrt(2))

  if ratio >= np.random.rand(1):

    x_values.append(x)
    x = x_rand
```

```python
y_values = np.histogram(x_values, bins = 100)[0] / np.histogram(x_values, bins = 100)[0].max()
new_x = np.histogram(x_values, bins = 100)[1]
```

```python
#Gaussian code
sigma = np.sqrt(2) #standard deviation
mu = 2 #mean
x = np.linspace(-5,8,10**4)
y = 1/(sigma*np.sqrt(2*np.pi))*np.exp((-1/2)*((x-mu)/sigma)**2)

y = y / y.max()
```

```python
plt.stairs(y_values,new_x, label = 'MCMC')
plt.plot(x,y, label = 'Gaussian fit')
plt.title('Comparing the MCMC to a Gaussian fit')
plt.ylabel('Normalized Counts')
plt.xlabel('x parameter bins')
plt.legend()
plt.show()
```

###Figure 1:

This plot compares how the MCMC sampler compares to the expected output, which is illustrated by the gaussian fit. Running the sampler for $10^5$ counts has knocked out the majority of the possible noise that could be seen, so the gaussian appears to be a great fit for the MCMC data.
