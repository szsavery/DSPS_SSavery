```python
import numpy as np
import pylab as pl
import astropy
from astropy import units as u

%pylab inline
```

```python
def square(l, ax):
  ax.plot([-1,1],[-1,-1],'k-')
  ax.plot([-1,1],[1,1],'k-')
  ax.plot([-1,-1],[-1,1],'k-')
  ax.plot([1,1], [-1,1],'k-')
```

```python
def circle(l, ax):
  theta = np.linspace(0, 360, 100)
  theta *= u.degree
  x = l * np.cos(theta.to(u.rad))
  y = l * np.sin(theta.to(u.rad))
  ax.plot(x, y, 'k-')
```

```python
fig, ax = pl.subplots(figsize=(5, 5))
square(1, ax=ax)
circle(1, ax=ax)
```

```python
def isinx(x,y,l):
  if np.sqrt(x**2 + y**2) <= l:
    return True

  return False
```

```python
np.random.seed(320)
counter = 0
N = 1000
for i in range(N):
  x = np.random.rand()* 2 - 1
  y = np.random.rand()* 2 - 1

  if isinx(x,y,1):
    counter +=1
```
